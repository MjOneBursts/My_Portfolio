```python
#Webscrapping Script
#The goal here is to create a CSV with Valorant weapon stats by rank, tier, and patch for competitive 

import requests 
from bs4 import BeautifulSoup
import pandas as pd 
import lxml 
base_url = 'https://www.metasrc.com/valorant/stats/weapons'

page = requests.get(base_url) #Test call, making sure we get a 200 response 
page
```

```python
import requests
import pandas as pd

patches = [1.14, 2.11, 3.12, 4.0, 4.01, 4.02, 4.03, 4.04, 4.05, 4.07, 4.08, 4.09, 4.10, 4.11, 5.0, 5.01, 5.03, 5.04, 5.05, 5.06, 5.07, 5.08, 5.09] # all available patches 
ranks = ['iron', 'silver', 'gold', 'platinum', 'diamond', 'immortal', 'radiant'] #all ranks in order of skill ascending 
base_url = 'https://www.metasrc.com/valorant/'
patch = '/stats/weapons?'
rank = 'ranks='
all_dfs = []
for p in patches : #Loop through all rank/patch combinations and build dataframe 
    for r in ranks: 
        url = base_url + str(p) + patch + rank + r 
        html = requests.get(url).content
        df_list = pd.read_html(html)
        df = df_list[0]
        df['Patch'] = p
        df['Rank'] = r 
        all_dfs.append(df)



print(all_dfs)
```

```python
weapons = pd.concat(all_dfs) #Create final dataframe 
weapons.head()
```
